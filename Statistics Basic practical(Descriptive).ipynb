# Measure of dispersion
1.variance

2.standard deviation

```python
import numpy as np
import statistics as st
print(st.variance(ages)) ## uses the sample variance formula to calculate variance
print(np.var(ages)) ## uses the population variance formuala to calculate variance
```

```python
def variance(data):
    n=len(ages)
    mean=sum(ages)/n
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/n ##divide  n for population variance and n-1 for sample variance
    return variance
    
```

```python
variance(ages)
```

```python
def sampolvar(data,dof=0):
    n=len(data)
    mean=sum(data)/n
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-dof)
    return variance
```

```python
populationvariance=sampolvar(ages,dof=0)
```

```python
import math as ma
ma.sqrt(populationvariance) ## standard deviation
```

# log normal,powerlaw distribution

```python
m,sigma=3,1 #mean and standard deviaton
s=np.random.lognormal(m,sigma,1000)
sns.histplot(s,kde=True)
plt.title("log normal,powerlaw distribution")
```

# converting log normal,powerlaw distribution into normal distribution

```python
sns.histplot(np.log(s),kde=True)
```
